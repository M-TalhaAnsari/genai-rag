## <b>Build a Multi-Agent Chatbot with AG2 (AutoGen) for Healthcare

```python
# importing required libraries
import warnings

# Suppress autogen and other deprecation/user warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings('ignore', category=UserWarning)

from autogen import ConversableAgent, GroupChat, GroupChatManager
from google import genai
from google.genai import chats
import logging

# Suppress warnings from autogen.oai.client
logging.getLogger("autogen.oai.client").setLevel(logging.ERROR)
```

When user interact following AI agents collabrate in realtimes, cross-validatind and optimizing their response to ensure accuracy and reliability<br>
1. Patient Agent
2. Sysmptom Analyzer Agent
3. Pharmacy Agent
4. Consultaion Advisor Agent

```python
import os
from dotenv import load_dotenv
load_dotenv()  # Load environment variables from .env file
llm_config = [
    {
        "model": "llama-3.3-70b-versatile",   # or any Groq-supported model
        "api_key": os.getenv("GROQ_API_KEY"),
        "base_url": "https://api.groq.com/openai/v1"
    }
]
```

<b>Creatign Conversable Agent

Conversable agent represents AI agents that can engage in conversations. Each agent has specifc role, defined by its system message. Use LLM configurations to process response.

```python
# create AI agents with defined roles
patient_agent = ConversableAgent(
    name="Patient",
    system_message="You describe symptoms and ask for medical help",
    llm_config=llm_config
)

diagnosis_agent = ConversableAgent(
    name="Diagnosis",
    system_message="You are a medical professional who provides diagnosis and advice based on the patient's symptoms",
    llm_config=llm_config   
)
pharmancy_agent = ConversableAgent(
    name="Pharmacy",
    system_message="You are a pharmacist who provides medication advice based on the diagnosis",
    llm_config=llm_config
)
consulation_agent = ConversableAgent(
    name="Consultation",
    system_message="You are a medical consultant who provides final advice and recommendations based on the diagnosis and pharmacy advice",
    llm_config=llm_config
)
```

<b>Group Chat

Manages a structured conversation between multiple AI agents.<br>
Ensures turn-based speaking using speaker selection method<br>
prevents infinite loop by using max_rounds

The GroupChat class structures the interaction between multiple ai agents, ensuring a logical conversation flow. 

```python
# Creating GroupChat for structured conversation
group_chat = GroupChat(
    agents = [patient_agent, diagnosis_agent, pharmancy_agent, consulation_agent],
    messages = [],
    max_round=5,
    speaker_selection_method="round_robin" )
```

```python
# creating GroupChatManager to manage the conversation
manager = GroupChatManager(name="manager", groupchat=group_chat)
```

```python
# Get patient and start consultation
symptoms = input("Please describe your symptoms: ")
response = manager.initiate_chat(manager, message=f"I am feeling {symptoms}. Can you help me?")
```
