```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
```

```python
import os
```

```python
model = ChatGoogleGenerativeAI(model="gemini-3.5-flash", api_key = os.getenv("GOOGLE_API_KEY"), temperature = 0) 
```

```python
prompt =  ChatPromptTemplate.from_messages([
    ("system", "You are world class technical document writer"),
    ("user", "{input}")
])

```

```python
output_parser = StrOutputParser()
```

```python
chain = prompt | model | output_parser

output = chain.invoke({'input':"Difference between langchain and langgraph"})

```

```python
print(output)
```
